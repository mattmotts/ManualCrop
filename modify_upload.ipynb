### Step 1
Pull a copied version of all images ending in '1x' into a newly created folder called 'upscale_to_10x' ... each 1x should retain their name


```python
import os
import shutil

# Replace these with the actual paths to your directories
source_dir = r'/Users/Matt/Pictures/MidJourney/Digital Products/Brushed by the Breeze/Images/Prelisting'
destination_dir = r'/Users/Matt/Pictures/MidJourney/Digital Products/Brushed by the Breeze/Images/upscale_to_10x'

# # Create the destination directory if it doesn't exist
# if not os.path.exists(destination_dir):
#     os.makedirs(destination_dir)

# Walk through the source directory
for root, dirs, files in os.walk(source_dir):
    for file in files:
        # Check if the file contains '1x' in its name
        if '1x' in file and file.endswith('.png'):
            # Construct full file paths
            source_file = os.path.join(root, file)
            destination_file = os.path.join(destination_dir, file)
            # Copy the file to the destination directory
            shutil.copy2(source_file, destination_file)
            print(f'Copied: {source_file} to {destination_file}')

print('Done!')
```

Moves images from source folder, to target folder, of which the target folder contains sub folders with a similar naming structure to the image that is in the source folder. Any images left behind will remain in the source folder.

```python
import os
import shutil

source_directory = r'/Users/Matt/Pictures/MidJourney/Digital Products/Brushed by the Breeze/Images/revamp/upscayl'
target_directory = r'/Users/Matt/Pictures/MidJourney/Digital Products/Brushed by the Breeze/Images/revamp'


def move_images(source_dir, target_dir):
    # Get list of images in the source directory
    source_images = [img for img in os.listdir(source_dir) if img.endswith('10x.png')]

    for img in source_images:
        # Get the base name of the image (without the '10x.png' part and the trailing underscore)
        base_name = img[:-7].rstrip('_')

        # Define the target subfolder path
        target_subfolder = os.path.join(target_dir, base_name)

        # Define the source image path
        source_img_path = os.path.join(source_dir, img)

        # Check if the target subfolder exists
        if os.path.isdir(target_subfolder):
            # Define the target image path
            target_img_path = os.path.join(target_subfolder, img)

            # Check if the image already exists in the target subfolder
            if os.path.exists(target_img_path):
                # If the image exists in the target subfolder, delete the source image
                os.remove(source_img_path)
            else:
                # If the image does not exist in the target subfolder, move it
                shutil.move(source_img_path, target_img_path)
        else:
            # If the target subfolder does not exist, skip moving the image
            print(f"Target subfolder '{target_subfolder}' does not exist for image '{img}'.")

# Example usage
source_dir = r'/Users/Matt/Pictures/MidJourney/Digital Products/Brushed by the Breeze/Images/upscale_to_10x/upscayl_png_realesrgan-x4plus_10x'
target_dir = r'/Users/Matt/Pictures/MidJourney/Digital Products/Brushed by the Breeze/Images/Listings'
move_images(source_dir, target_dir)
```

create a new folder for each unique image name

```python
import os
import shutil

# Define source directory
source_directory = r'/Users/Matt/Pictures/MidJourney/Digital Products/Brushed by the Breeze/Images/revamp'

# Create a set to store unique beginnings of filenames
unique_beginnings = set()

# Iterate through each file in the source directory
for filename in os.listdir(source_directory):
    if os.path.isfile(os.path.join(source_directory, filename)):
        # Extract the unique beginning part of the filename
        unique_beginning = filename.rsplit('_', 1)[0]
        unique_beginnings.add(unique_beginning)

# Create folders and move files
for beginning in unique_beginnings:
    folder_path = os.path.join(source_directory, beginning)
    
    # Create the folder if it doesn't exist
    if not os.path.exists(folder_path):
        os.makedirs(folder_path)
    
    # Move files to the respective folders
    for filename in os.listdir(source_directory):
        if os.path.isfile(os.path.join(source_directory, filename)) and filename.startswith(beginning):
            shutil.move(os.path.join(source_directory, filename), os.path.join(folder_path, filename))
            print(f'Moved: {filename} -> {folder_path}')

print('All files have been organized into their respective folders.')
```

Here is a Python script that will walk through a directory and its subdirectories, find images with filenames containing "12x18" or "24x36", apply a watermark to the center of each image, and save the modified images with the naming scheme you specified:

```python
import os
from PIL import Image, ImageDraw

def add_frame_outline(input_image_path, output_image_path, outline_color, outline_width=10):
    original_image = Image.open(input_image_path).convert("RGBA")
    width, height = original_image.size

    # Create an image for the outline
    outline_image = Image.new("RGBA", (width, height), (0, 0, 0, 0))
    draw = ImageDraw.Draw(outline_image)

    # Draw the outline rectangle
    draw.rectangle(
        [(outline_width // 2, outline_width // 2), (width - outline_width // 2, height - outline_width // 2)],
        outline=outline_color,
        width=outline_width
    )

    # Combine the original image with the outline
    combined_image = Image.alpha_composite(original_image, outline_image)
    combined_image = combined_image.convert("RGB")  # Remove alpha for saving in jpg format

    combined_image.save(output_image_path)

def process_images(directory):
    for root, _, files in os.walk(directory):
        for file in files:
            if "12x18" in file:
                outline_color = (255, 0, 0, 255)  # Red
                file_path = os.path.join(root, file)
                base, ext = os.path.splitext(file)
                new_filename = f"{base}WM{ext}"
                output_path = os.path.join(root, new_filename)
                add_frame_outline(file_path, output_path, outline_color)
            elif "24x36" in file:
                outline_color = (0, 255, 0, 255)  # Green
                file_path = os.path.join(root, file)
                base, ext = os.path.splitext(file)
                new_filename = f"{base}WM{ext}"
                output_path = os.path.join(root, new_filename)
                add_frame_outline(file_path, output_path, outline_color)

# Example usage
directory = r'/Users/Matt/Pictures/MidJourney/Digital Products/Brushed by the Breeze/Images/revamp'
process_images(directory)
```

```python
import os

def delete_watermarked_images(directory):
    for root, _, files in os.walk(directory):
        for file in files:
            if file.endswith("WM.png"):
                file_path = os.path.join(root, file)
                try:
                    os.remove(file_path)
                    print(f"Deleted: {file_path}")
                except OSError as e:
                    print(f"Error deleting {file_path}: {e}")

# Example usage
directory = r'/Users/Matt/Pictures/MidJourney/Digital Products/Brushed by the Breeze/Images/revamp'
delete_watermarked_images(directory)
```
